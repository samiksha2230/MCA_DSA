```python
# Two-dimensional array using Python list
rows = int(input("Enter number of rows: "))
cols = int(input("Enter number of columns: "))

matrix = []

# Insertion
for i in range(rows):
    row = []
    for j in range(cols):
        element = int(input(f"Enter element at position ({i},{j}): "))
        row.append(element)
    matrix.append(row)

# Traversal
print("2D Array (Matrix):")
for i in range(rows):
    for j in range(cols):
        print(matrix[i][j], end=" ")
    print()
```
