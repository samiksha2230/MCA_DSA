```python
#ENQUEUE AND DEQUEUE
class Queue:
    def __init__(self):
        self.queue = []

    # Enqueue (Insert element at end)
    def enqueue(self, item):
        self.queue.append(item)
        print(item, "enqueued")

    # Dequeue (Remove element from front)
    def dequeue(self):
        if len(self.queue) == 0:
            print("Queue is empty")
        else:
            item = self.queue.pop(0)
            print(item, "dequeued")

    # Display queue
    def display(self):
        print("Queue:", self.queue)


# Example usage
q = Queue()

q.enqueue(10)
q.enqueue(20)
q.enqueue(30)

q.display()

q.dequeue()
q.display()
```
