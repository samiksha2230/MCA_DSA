```python
# One-dimensional array using Python list
arr = []

# Input size of array
n = int(input("Enter number of elements: "))

# Insertion
for i in range(n):
    element = int(input("Enter element: "))
    arr.append(element)

# Traversal
print("Array elements are:")
for i in range(len(arr)):
    print("Index", i, "=", arr[i])
```
