```python

#postfix expression using stack
def evaluate_postfix(expression):
    stack = []
    
    for char in expression:
        if char.isdigit():
            stack.append(int(char))
        else:
            b = stack.pop()
            a = stack.pop()
            
            if char == '+':
                result = a + b
            elif char == '-':
                result = a - b
            elif char == '*':
                result = a * b
            elif char == '/':
                result = a / b
            
            stack.append(result)
    
    return stack.pop()

expression = "23+3+4*"
print(evaluate_postfix(expression))
```
