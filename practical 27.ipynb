```python
import sys

cost = [
    [9, 2, 7],
    [6, 4, 3],
    [5, 8, 1]
]

n = len(cost)
min_cost = sys.maxsize

def branch_bound(worker, current_cost, assigned):
    global min_cost

    if worker == n:
        min_cost = min(min_cost, current_cost)
        return

    if current_cost >= min_cost:
        return

    for job in range(n):
        if not assigned[job]:
            assigned[job] = True
            branch_bound(worker + 1,
                         current_cost + cost[worker][job],
                         assigned)
            assigned[job] = False

assigned = [False] * n

branch_bound(0, 0, assigned)

print("Minimum Cost =", min_cost)
```
